***Create a function to store a labled data in image folder of different different persons***

```python
facedetect = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
count = 0
nameid = str(input("Enter your name: ")).lower()
# Path of your folder
path = 'C:\\Users\\shard\\810-Ethans-tech\\Computer vision\\image\\'+nameid

isExist = os.path.exists(path)
# name of person which face you take
if isExist:
    print("Name already taken!")
    nameid = str(input("Enter your name again: "))
else:
    os.makedirs(path)
```

**Capture the 500 faces**

```python
video = cv2.VideoCapture(0)
if not video.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = video.read()
    if not ret:
        break  # Break the loop if frame reading fails
    faces = facedetect.detectMultiScale(frame,1.3,5)
    for x,y,w,h in faces:
        count = count+1
        name = './image/'+nameid+'/'+str(count)+'.jpg'
        print('Create images............'+name)
        cv2.imwrite(name, frame[y:y+h,x:x+w])
        cv2.rectangle(frame, (x,y), (x+w, y+h),(0,255,0), 3)
    cv2.imshow('WindowFrame', frame)
    cv2.waitKey(1)
    if count>500:
        break

video.release()
cv2.destroyAllWindows()
```

# Code for encode the images which we make in above line for comparision when camera open and show some one face

```python
video = cv2.VideoCapture(0)
if not video.isOpened():
    raise IOError("Cannot open webcam")

# Load known face encodings and names
known_face_encodings = []
known_face_names = []

# Load previously saved images and encode them
base_path = "C:\\Users\\shard\\810-Ethans-tech\\Computer vision\\image"

# Walk through all subdirectories and files to get images
for root, dirs, files in os.walk(base_path):
    for file in files:
        file_path = os.path.join(root, file)

        # Skip non-image files or system files like .ipynb_checkpoints
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            try:
                # Load the image and generate face encodings
                image = face_recognition.load_image_file(file_path)
                encodings = face_recognition.face_encodings(image)
                
                # If face encodings are found, add the encoding and name
                if encodings:
                    print(f"Encoding generated for: {file}")
                    # Add encoding for the first face found in the image
                    known_face_encodings.append(encodings[0])
                    # Extract the name from the folder (assuming the folder is named after the person)
                    person_name = os.path.basename(root)  # folder name is the person's name
                    known_face_names.append(person_name)
                else:
                    print(f"No face found in image: {file}")
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
        else:
            print(f"Skipped non-image file: {file}") 
```

***Code for face recognation when webcam open***

```python

# Start capturing video
while True:
    ret, frame = video.read()
    if not ret:
        break  # Break if frame capture fails

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
        face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
        name = "Unknown"

        if matches:
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]:
                name = known_face_names[best_match_index]

        # Draw rectangle and label
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.putText(frame, name, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    cv2.imshow("Face Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit
        break

video.release()
cv2.destroyAllWindows()
```
